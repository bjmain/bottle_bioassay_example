```python
import pandas as pd
%matplotlib inline


#treatment DE genes
m = pd.read_table("mortality.txt")
print(m)
#trt_diff.columns = ['Acol_gene', 'logFC', 'AveExpr', 't', 'P.Value', 'adj.P.Val', 'B']
```

```python
%matplotlib inline
from itertools import cycle, islice

my_colors = list(islice(cycle(['b', 'b','b', 'b', 'g', 'g', 'g', 'g', 'r', 'r', 'r', 'r']), None, 12))

ax = m['mean_dead_%'].plot(alpha=1,fontsize=24,color = my_colors,kind='bar', figsize=(15, 10), yerr=m['SE'], rot=70 )

ax.set_ylabel("Mean mortality @ 24 hours (%)",fontsize=24)

ax.set_xticklabels(m['geno_treatment'])
ax.set_xlabel("",fontsize=24)
```
